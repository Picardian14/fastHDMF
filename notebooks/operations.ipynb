```python
import os 
import shutil
#sub-HC001_ses-01_space-dwinative_atlas-glasser360_desc-sc.txt
for sub_folder in os.listdir("/network/iss/cohen/data/Ivan/VirtualFingerprint/MICs_Dataset/MICs/derivatives/micapipe/micapipe"):
    if sub_folder.startswith("sub-"):
        print(f"Processing subject folder: {sub_folder}")
        for ses_folder in os.listdir(f"/network/iss/cohen/data/Ivan/VirtualFingerprint/MICs_Dataset/MICs/derivatives/micapipe/micapipe/{sub_folder}"):            
            if ses_folder.startswith("ses-"):
                print(f"  Processing session folder: {ses_folder}")
                for file in os.listdir(f"/network/iss/cohen/data/Ivan/VirtualFingerprint/MICs_Dataset/MICs/derivatives/micapipe/micapipe/{sub_folder}/{ses_folder}/dwi"):
                    if file.endswith("glasser360_desc-sc.txt") and "space-dwinative" in file:
                        old_path = f"/network/iss/cohen/data/Ivan/VirtualFingerprint/MICs_Dataset/MICs/derivatives/micapipe/micapipe/{sub_folder}/{ses_folder}/dwi/{file}"
                        new_file = file.split("_")[0] + ".csv"
                        new_path = f"/network/iss/home/ivan.mindlin/Repos/fastHDMF/data/raw/MICA_SCs/glasser360/{new_file}"
                        shutil.copy2(old_path, new_path)
                        #os.rename(old_path, new_path)
                        print(f"Renamed: {old_path} to {new_path}")
                        
                        
```
